```python
import pandas as pd
import datetime
```

```python
# Lets create a variable with dates, and then from dates we extract the weekdays
# create a list of dates having 20 days difference from today
ds_base = datetime.datetime.today()
ds_date_list = [ds_base - datetime.timedelta(days=x) for x in range(0,20)]
ds = pd.DataFrame(ds_date_list)
ds.columns=['day']
```

```python
ds.head()
```

```python
# Lets extract the week days
ds['weekdays'] = ds['day'].dt.day_name()
```

```python
ds.head()
```

```python
# Now lets perform ordinal encoding/ label encoding
# Here we will map sunday to 1, monday to 2, ... , saturday to 7
# This ensures that weekend will get higher weights, which is what most people want
weekday_dict = {
    'Sunday':1,
    'Monday':2,
    'Tuesday':3,
    'Wednesday':4,
    'Thursday':5,
    'Friday': 6,
    'Saturday':7
}

ds['weekdays_ordinal'] = ds['weekdays'].map(weekday_dict)
```

We can see that we have successfully performed the ordinal encoding
