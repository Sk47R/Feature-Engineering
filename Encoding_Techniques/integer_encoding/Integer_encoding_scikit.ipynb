```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
# from feature_engine.encoding import OrdinalEncoder
```

```python
ds = pd.read_csv('mercedesbenz.csv', usecols=['X1', 'X2','X3', 'X4','X5', 'X6'])
```

```python
ds.head()
```

```python
# lets find out how many unique categories are present in each columns
for column in ds.columns:
  print(f"Column {column} has {len(ds[column].unique())} categories")
```

```python
# Create and fit the encoder

le = LabelEncoder()
le.fit(ds['X2'])
```

```python
# Observing Unique Classes
le.classes_
```

```python
# Now lets transform and view the result
ds['X2'] = le.transform(ds['X2'])

ds.head()
```

We can see that the feature column X2 has been integer encoded.

Note: LabelEncoder works one variable at the time. However there is a way to automate this for all the categorical variables. 



```python
from collections import defaultdict
```

```python
d = defaultdict(LabelEncoder)
```

```python
# Encoding the variable
ds_transformed = ds.apply(lambda x: d[x.name].fit_transform(x))
```

```python
ds_transformed.head()
```

```python
# If we want to bring back the original labels, we can inverse transform the matrix

# Inverse the encoded
tmp = ds_transformed.apply(lambda x: d[x.name].inverse_transform(x))
tmp.head()
```
