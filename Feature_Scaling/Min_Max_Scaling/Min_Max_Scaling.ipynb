For practising standardization, I will use boston dataset available in sklearn

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

# dataset
from sklearn.datasets import load_boston
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import MinMaxScaler
```

```python
# load the the Boston House price data
boston_dataset = load_boston()

# Create the dataframe
ds = pd.DataFrame(boston_dataset.data,
                      columns=boston_dataset.feature_names)

# add target
ds['MEDV'] = boston_dataset.target
```

```python
# lets check the data's
ds.head()
```

```python
# let's have a look at the main statistical parameters of the variables
# to get an idea of the feature magnitudes

ds.describe()
```

The different variables shows different value ranges, mean, max, min, standard deviations, etc. In other words, they show different magnitudes or scales. Note, how the minimum and maximum values are are quite different in the different variables.

When performing min-max scaling on the data set, we need to first identify the minimum and maximum values of the variables. These parameters need to be learned from the train set, stored, and then used to scale test and future data. Thus, we will first divide the data set into train and test.



```python
# Splitting the train and test data
X_train, X_test, y_train, y_test = train_test_split(ds.drop('MEDV', axis=1),
                                                    ds['MEDV'],
                                                    test_size=0.3,
                                                    random_state=0)

# print the shape
X_train.shape, X_test.shape
```

### Mean Max Scaling
The MinMaxScaler from scikit-learn subtracts the minimum value and re-scales the the value range, rendering the variable values between 0 and 1.

```python
# set up the scaler
scaler = MinMaxScaler()

# fit the scaler to the train set, it will learn the parameters
scaler.fit(X_train)

# transform train and test sets
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# let's print the maximum values of the features
print(scaler.data_max_)
```

```python
# let's print the minimum values of the features
print(scaler.min_)
```

```python
# lets print the range i.e (max - mean) of each features

print(scaler.data_range_)
```

```python
# let's transform the returned NumPy arrays to dataframes

X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns)
```

```python
# let's have a look at the original training dataset: mean and standard deviation
# np.round to reduces the number of decimals to 1.

np.round(X_train.describe(), 1)
```

```python
# Now, let's have a look at the scaled training dataset: mean and standard deviation
# np.round to reduces the number of decimals to 1.

np.round(X_train_scaled.describe(), 1)
```

We can see that, the minimum and maximum values for all the variables are 0 and 1, respectively. The mean is not centered at zero, and the variance changes.



```python
# let's compare the variable distributions before and after scaling

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['RM'], ax=ax1)
sns.kdeplot(X_train['LSTAT'], ax=ax1)
sns.kdeplot(X_train['CRIM'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['RM'], ax=ax2)
sns.kdeplot(X_train_scaled['LSTAT'], ax=ax2)
sns.kdeplot(X_train_scaled['CRIM'], ax=ax2)
plt.show()
```

We can see that the values are now capped at 1, but the distributions are not centered.


```python
# let's compare the variable distributions before and after scaling for some 
# other features

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(12, 5))

# before scaling
ax1.set_title('Before Scaling')
sns.kdeplot(X_train['AGE'], ax=ax1)
sns.kdeplot(X_train['DIS'], ax=ax1)
sns.kdeplot(X_train['NOX'], ax=ax1)

# after scaling
ax2.set_title('After Standard Scaling')
sns.kdeplot(X_train_scaled['AGE'], ax=ax2)
sns.kdeplot(X_train_scaled['DIS'], ax=ax2)
sns.kdeplot(X_train_scaled['NOX'], ax=ax2)
plt.show()
```
